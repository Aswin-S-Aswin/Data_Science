#### Level 7 - Take this program to run as a function when it is called

```python
import random

print("Welcome to the Number Guessing Game!")
print("I'm thinking of a number between 1 and 100")

secret_num = random.randint(1,100)

for i in range(8):
  user_input= int(input(f"{i}. Enter a number : "))
  if user_input < 1 or user_input > 100:
    print("Invalid input. Please enter a number between 1 and 100.")
  elif user_input > secret_num:
    print("Too high")
  elif user_input < secret_num:
    print("Short broo")
  else:
    print("Congratulation! your guess is right!")
    break

if secret_num == user_input:
  print("Super bro")
else:
  print("Ada Poda")
  print(f"Output vantu : {secret_num}")
```

#### Level 6 - How to handle invalid characters

```python
import random

print("Welcome to the Number Guessing Game!")
print("I'm thinking of a number between 1 and 100")

secret_num = random.randint(1,100)

for i in range(8):
  try:
    user_input= int(input(f"{i}. Enter a number : "))
    if user_input < 1 or user_input > 100:
      print("Invalid input. Please enter a number between 1 and 100.")
    elif user_input > secret_num:
      print("Too high")
    elif user_input < secret_num:
      print("Short broo")
    else:
      print("Congratulation! your guess is right!")
      break
  except ValueError:
    print("Invalid input. Please enter a valid number.")

if secret_num == user_input:
  print("Super bro")
else:
  print("Ada Poda")
  print(f"Output vantu : {secret_num}")
```
